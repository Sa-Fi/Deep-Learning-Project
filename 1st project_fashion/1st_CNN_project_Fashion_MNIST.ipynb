````python
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns 
import tensorflow as tf
import keras 
````

Load Dataset from keras dataset 
Here we use fashion mnist dataset

````python
(x_train,y_train),(x_test,y_test)=tf.keras.datasets.fashion_mnist.load_data()
````

````python
print("x train shaped",x_train.shape)
print("y train shaped",y_train.shape)
print("x test shaped",x_test.shape)
print("y test shaped",y_test.shape)
````

````python
x_train
````

````python
x_train[0]
````

````python
y_train[0]
````

````python
class_labels =["T-shirt/top","Trouser","Pullover","Dress","Coat"
               ,"Sandal","Shirt","Sneaker","Bag","Ankle boot"]
class_labels
````

show image using matplotlib 

````python

plt.imshow(x_train[0],cmap="Greys")
````

````python
plt.figure(figsize=(8,8))
j=1
for i in range(25):
  plt.subplot(5,5,j); j+=1
  plt.imshow(x_train[i], cmap="Greys")
  plt.axis('off') # off the axis
  plt.title('{} / {}'.format(class_labels[y_train[i]], y_train[i]))
````

Change Dimension

````python
x_train.ndim
````

````python
x_train = np.expand_dims(x_train,-1)
x_test = np.expand_dims(x_test,-1)
x_train.ndim
````

````python
x_train.shape
````

Features Scaling

````python
x_train = x_train/255 
# 0 theke 255 er modde value gula jehto boro train korte somoi lage so ei ketre 255 die divide kore value gula 0 theke 1 er modde nie asa hoi jate training aro efficient hoi 
x_test = x_test/255
````

````python
x_train[0]
````

````python
# training er somoi train data input korte hoi sathe validation data 
# validation data means emn kicu data jeta model train howar somoi train data gula ke valid kore 
# jkn model train hoi jabe tkn test data die value predict kora lagbe 
# so akn amder theke test data ache ,, to amder validation data prepare korte hobe 
````

split dataset 


````python
from sklearn.model_selection import train_test_split
x_train,x_val,y_train,y_val = train_test_split(x_train,y_train,test_size=0.2,random_state=2020)
````

````python
print("x train shaped",x_train.shape)
print("y train shaped",y_train.shape)
print("x val shaped",x_val.shape)
print("y val shaped",y_val.shape)
````

CNN model building

````python
model = keras.models.Sequential([
    keras.layers.Conv2D(filters = 32,kernel_size=3,strides=(1,1),padding="valid",activation="relu",input_shape=(28,28,1)),
    keras.layers.MaxPooling2D(pool_size=(2, 2)),
    keras.layers.Flatten(),
    keras.layers.Dense(units=128,activation="relu"),
    keras.layers.Dense(units=10,activation="softmax")
])
````

````python
model.summary()
````

````python
model.compile(optimizer="adam",loss="sparse_categorical_crossentropy",metrics=['accuracy'])
````

````python
model.fit(x_train,y_train,epochs=10,batch_size=512,verbose=1,validation_data=(x_val,y_val))
````

Test and evaluate the model

````python
model.predict(np.expand_dims(x_test[0],axis=0)).round(2)
````

````python
np.argmax(model.predict(np.expand_dims(x_test[0],axis=0)).round(2))
````

````python
y_test[0]
````

````python
y_pred = model.predict(x_test).round(2)
y_pred
````

````python
model.evaluate(x_test,y_test)
````

Visualize output

````python
 plt.figure(figsize=(30,30))
 j=1
 for i in range(25):
   plt.subplot(5,5,j)
   j+=1
   plt.imshow(x_test[i],cmap='Greys')
 #  plt.title('Actual = {} / {} \nPredictedf = {} / {}'.format(class_labels[y_test[i]],y_test[i],class_labels[np.argmax(y_pred[i])],np.argmax[y_pred[i]]))
   plt.title('Actual = {} / {} \nPredicted = {} / {}'.format(class_labels[y_test[i]], y_test[i], class_labels[np.argmax(y_pred[i])],np.argmax(y_pred[i])))
   plt.axis(False)
````

**Confusion Matrix**





````python
from sklearn.metrics import confusion_matrix
plt.figure(figsize=(16,9))
y_pred_labels = [np.argmax(label) for label in y_pred]
cm =confusion_matrix(y_test,y_pred_labels)

````

````python
# show confusion matrix 
sns.heatmap(cm,annot=True,fmt='d',xticklabels=class_labels,yticklabels=class_labels)
````

````python
# precision = true_positive/(true_positive + false_positive)
# Recall = true_positive/ (true_positive + false_negative)
from sklearn.metrics import classification_report
cr = classification_report(y_test,y_pred_labels,target_names = class_labels)
print(cr)
````
